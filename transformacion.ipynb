# Transformacion de Datasets

```
import pandas as pd
```

## Dataframes a partir de CSV's

```
df_amazon = pd.read_csv (r'amazon_prime_titles-score.csv', sep = ',', header = 0)
df_amazon
```

```
df_disney = pd.read_csv(r'disney_plus_titles-score.csv', sep = ',', header = 0)
df_disney
```

```
df_hulu = pd.read_csv(r'hulu_titles-score (2).csv', sep = ',', header = 0)
df_hulu
```

```
df_netflix = pd.read_csv(r'netflix_titles-score.csv', sep = ',', header = 0)
df_netflix
```

### Generar campo id: 

Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = as123).

Mediante el método apply y funcion lambda agrego la letra inicial de la plataforma correspondiente a su id del campo 'show_id' y de esta manera genero su id unico e irrepetible entre los demás id's de las demás plataformas. Mismos valores que irán en la el campo 'id'. 

```
df_amazon['id'] = df_amazon['show_id'].apply(lambda x: 'a' + x)
df_amazon
```

```
df_disney ['id'] = df_disney['show_id'].apply(lambda x: 'd' + x)
df_disney
```

```
df_hulu['id'] = df_hulu['show_id'].apply(lambda x: 'h' + x)
df_hulu
```

```
df_netflix['id'] =df_netflix['show_id'].apply(lambda x: 'n' + x)
df_netflix
```

### Valores nulos del campo 'rating'

Los valores nulos del campo rating los reemplazo por el string "g" el cuál corresponde al maturity rating: "General for al audiences". Esto mediante el método fillna().

```
df_amazon['rating'].fillna('g',inplace=True)
df_amazon
```

```
df_disney['rating'].fillna('g',inplace=True)
df_disney
```

```
df_hulu['rating'].fillna('g', inplace=True)
df_hulu
```

```
df_netflix['rating'].fillna('g', inplace=True)
df_netflix
# en netflix no es necesaria esta transformación porque no había datos faltantes para este campo.
```

### Formato de fechas

Los campos correspondientes a fechas en este caso 'date_added' serán transformados a un tipo de dato de fecha mediante metodo pd.to_datetime(). 

```
df_amazon['date_added'] = pd.to_datetime(df_amazon.date_added, infer_datetime_format=True)
df_amazon
```

```
df_disney['date_added'] = pd.to_datetime(df_disney.date_added, infer_datetime_format=True)
df_disney
```

```
df_hulu['date_added'] = pd.to_datetime(df_hulu.date_added, infer_datetime_format=True)
df_hulu
```

```
df_netflix['date_added'] = pd.to_datetime(df_netflix.date_added, infer_datetime_format=True)
df_netflix
```

### Minusculas

Todo campo con valores de tipo string seran normalizados a letras minusculas, aplicando una funcion lambda mediante metodo apply.

```
df_amazon = df_amazon.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)
df_amazon
```

```
df_disney = df_disney.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)
df_disney 
```

```
df_hulu = df_hulu.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)
df_hulu 
```

```
df_netflix = df_netflix.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)
df_netflix
```

### Campo "duration" a dos campos

El campo "duration"  lo convierto a dos campos: duration_int y duration_type, mediante metodo split().

duration_int contendra valores  integer que indica la cantidad ya sea de minutos o de temporadas segun sea el caso; y duration_type un string indicando la unidad de medición de duración: min (minutos) o season (temporadas).

```
df_amazon[['duration_int', 'duration_type']] = df_amazon['duration'].str.split(expand=True)
df_amazon['duration_int'] = df_amazon['duration_int'].astype('int64')
df_amazon
```

```
df_disney[['duration_int', 'duration_type']] = df_disney['duration'].str.split(expand=True)
df_disney['duration_int'] = df_disney['duration_int'].astype('int64')
df_disney
```

```
df_hulu[['duration_int', 'duration_type']] = df_hulu['duration'].str.split(expand=True)
df_hulu['duration_int'] = df_hulu['duration_int'].astype(float).astype('Int64')
df_hulu
```

```
df_netflix[['duration_int', 'duration_type']] = df_netflix['duration'].str.split(expand = True)
df_netflix['duration_int'] = df_netflix['duration_int'].astype(float).astype('Int64')
df_netflix
```

### Normalizo "seasons" a "season" en duration_type

Normalizacion mediante metodo apply y funciones lambda, aunque tambien puede ser usado el metodo replace()

```
df_amazon['duration_type'] = df_amazon['duration_type'].apply(lambda x: 'season' if x == 'seasons'else x)
df_amazon
```

```
df_disney['duration_type'] = df_disney['duration_type'].apply(lambda x: 'season' if x == 'seasons' else x)
df_disney
```

```
df_hulu['duration_type'] = df_hulu['duration_type'].apply(lambda x: 'season' if x == 'seasons' else x)
df_hulu
```

```
df_netflix['duration_type'] = df_netflix['duration_type'].apply(lambda x: 'season' if x == 'seasons' else x)
df_netflix
```

### Concateno dataframes a uno solo

Esto para facilitar las consultas que requieran una busqueda global, es decir entre los datos de todas las plataformas de streaming.

```
df = pd.concat([df_amazon, df_disney, df_hulu, df_netflix])
```

```
df
```

Por ultimo, exporto los datasets en formato .csv con los datos ya transformados a partir de los dataframes. 

```
df_amazon.to_csv('amazon.csv', sep = ',', index = False)
df_disney.to_csv('disney.csv', sep = ',', index = False)
df_hulu.to_csv('hulu.csv', sep = ',', index = False)
df_netflix.to_csv('netflix.csv' , sep = ',', index = False)
```
